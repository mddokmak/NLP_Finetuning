```python
from transformers import pipeline,  set_seed
from transformers import AutoTokenizer

device_map = {"": torch.cuda.current_device()} if torch.cuda.is_available() else None

set_seed(42)

# Define model kwargs
model_kwargs = dict(
    torch_dtype="auto",
    use_cache=False,
    device_map=device_map
)

model = AutoModelForCausalLM.from_pretrained('openai-community/gpt2', **model_kwargs)

tokenizer = AutoTokenizer.from_pretrained(reference_model_path)
# Define device map for the model

generator = pipeline('text-generation', model=model, tokenizer=tokenizer)

question = 'Answer the following multiple-choice questions by selecting the correct letter (A, B, C, ...).\n\nQuestion: What is the capital of France?\n\nOptions:\nA) Berlin\nB) Madrid\nC) Paris\nD) Rome\n\nAnswer: C\n\nQuestion: What is 2 + 2?\n\nOptions:\nA) 3\nB) 2\nC) 4\nD) 7\n\nAnswer: '
print(question)

prediction = generator(question, truncation = True, max_length = 256, num_return_sequences=1)

print(prediction)
```
